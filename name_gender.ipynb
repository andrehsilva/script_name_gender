```python
import pandas as pd
from genderbr import get_gender

# Caminho para o arquivo Excel
caminho_arquivo_excel = '/content/drive/MyDrive/Colab Notebooks/Conexia/name_gender_ias/name_gender.xlsx'

# Lendo o arquivo Excel
try:
    df = pd.read_excel(caminho_arquivo_excel)
except FileNotFoundError:
    print("Arquivo não encontrado.")
    exit()

# Verifica se a coluna 'nome_completo' está presente no DataFrame
if 'nome' not in df.columns:
    print("A coluna 'nome' não está presente no arquivo Excel.")
    exit()

# Extrair o primeiro nome e salvar em uma nova coluna 'primeiro_nome'
df['primeiro_nome'] = df['nome'].str.split().str.get(0)

df.head()
```

```python
def identificar_genero(nome):
    # Obtém o gênero do nome
    genero = get_gender(nome)
    return genero

# Verifica se a coluna 'nome' está presente no DataFrame
if 'primeiro_nome' not in df.columns:
    print("A coluna 'primeiro_nome' não está presente no arquivo Excel.")
    exit()

# Aplica a função identificar_genero para cada nome na coluna 'nome' e adiciona o resultado em uma nova coluna 'genero'
df['genero'] = df['primeiro_nome'].apply(identificar_genero)


df.head()
```

```python
df_nomes = df.copy()
```

```python
# Atualiza os valores na coluna 'genero' usando uma função lambda e uma expressão condicional ternária
df_nomes['genero'] = df_nomes['genero'].apply(lambda x: 'Masculino' if x == 'M' else ('Feminino' if x == 'F' else 'Desconhecido'))
df_nomes.head()
```

```python
df_nomes['genero_desconhecido_para_masculino'] = df_nomes['genero'].apply(lambda x: 'Masculino' if x == 'Desconhecido' else ('Feminino' if x == 'Feminino' else 'Masculino'))
```
